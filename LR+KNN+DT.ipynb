```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
training_data = data[['texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
        'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
X, x_test, Y, y_test= train_test_split(training_data, output_data, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier()
log = LogisticRegression()
tree = DecisionTreeClassifier()
```

```python
def model_training ():
    #KNN Model
    knn.fit(X,Y) #training
    knn_pridict = knn.predict(x_test)
    knn_accuracy = accuracy_score(knn_pridict,y_test)
    
    #LR
    log.fit(X,Y)
    log_pridict = log.predict(x_test)
    log_accuracy = accuracy_score(log_pridict,y_test)
    
    
    #DT
    tree.fit(X,Y)
    tree_pridict = tree.predict(x_test)
    tree_accuracy = accuracy_score(tree_pridict,y_test)
    
    print(knn_accuracy*100 ,"%")
    print(log_accuracy*100,"%")
    print(tree_accuracy*100,"%")
```

```python
model_training()
```
